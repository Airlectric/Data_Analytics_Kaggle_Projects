### Feature Selection

When building a machine learning model, selecting the appropriate features from the dataset is crucial for the model's performance and interpretability. Here, we select relevant features to be used as input variables for our model.

#### Selected Features:
- `InvocationDelay`: Time delay between the invocation and response of a function.
- `ResponseDelay`: Delay in the response time of a function.
- `FunctionDuration`: Duration of the function execution.
- `maxcpu`: Maximum CPU usage.
- `avgcpu`: Average CPU usage.
- `p95maxcpu`: 95th percentile of the maximum CPU usage.

#### Additional Features:
- `SubmitTime`: Timestamp of when the request was submitted.
- `RTT`: Round-Trip Time, i.e., the time taken for a signal to travel to a destination and back.
- `ActiveFunctionsAtRequest`: Number of active functions at the time of request.
- `ActiveFunctionsAtResponse`: Number of active functions at the time of response.

These features are chosen based on their potential relevance to the detection of Denial-of-Service (DoW) attacks, assuming that the 'bot' column represents the binary label indicating the presence of such attacks.


```python
# Select relevant features
selected_features = ['InvocationDelay', 'ResponseDelay', 'FunctionDuration', 'maxcpu', 'avgcpu', 'p95maxcpu']
additional_features = ['SubmitTime', 'RTT', 'ActiveFunctionsAtRequest', 'ActiveFunctionsAtResponse']
X = dataset[selected_features + additional_features]
y = dataset['bot']  
```

### Feature Engineering

Feature engineering involves transforming raw data into a format that is suitable for machine learning algorithms. Here, we perform feature engineering on the dataset to extract additional time-based features from the timestamp column.

#### Steps:

1. **Timestamp Conversion**: Convert the timestamp column to datetime format to facilitate time-based feature extraction.

2. **Time-Based Features Extraction**:
   - `HourOfDay`: Extract the hour of the day from the timestamp.
   - `DayOfWeek`: Extract the day of the week from the timestamp.

These time-based features can provide valuable insights into the temporal patterns and trends present in the data, which may be useful for detecting patterns related to Denial-of-Service (DoW) attacks.


```python
X = dataset.copy()

# Convert timestamp to datetime format
X['timestamp'] = X['timestamp'].apply(lambda x: pd.to_datetime(x, format='mixed', utc=True))

# Extract time-based features
X['HourOfDay'] = X['timestamp'].dt.hour
X['DayOfWeek'] = X['timestamp'].dt.dayofweek
```

### Interaction Features

Interaction features are created by combining or interacting two or more existing features to capture potential relationships or interactions between them. Here, we create interaction features using specific columns from the dataset.

#### Steps:

1. **Interaction1**: Multiply the 'InvocationDelay' column by the 'ResponseDelay' column to create a new interaction feature, 'Interaction1'.
   
2. **Interaction2**: Divide the 'maxcpu' column by the 'avgcpu' column to create another interaction feature, 'Interaction2'.

These interaction features aim to capture potential synergies or dependencies between the specified columns, which may help improve the model's performance by incorporating additional information about the data.


```python
# 2. Interaction features
X['Interaction1'] = X['InvocationDelay'] * X['ResponseDelay']
X['Interaction2'] = X['maxcpu'] / X['avgcpu']
```

### Statistical Features

Statistical features provide insights into the distribution and variability of specific attributes across different groups or categories. These features are calculated based on group-wise operations on the dataset, particularly focusing on the behavior of certain attributes in relation to unique IP addresses.

1. **RTT_mean**: This feature represents the average Round-Trip Time (RTT) observed for each unique IP address. RTT is a crucial metric in network communications, indicating the time taken for a signal to travel from the sender to the receiver and back. By computing the mean RTT for each IP address, we gain an understanding of the typical communication latency experienced by devices associated with that IP.

2. **RTT_std**: The standard deviation of RTT provides a measure of the variability or dispersion in RTT values across different instances associated with the same IP address. A higher standard deviation suggests greater variability in communication latencies, which could indicate fluctuating network conditions or diverse types of traffic handled by the IP.

3. **InvocationDelay_mean**: This feature denotes the average Invocation Delay observed for each unique IP address. Invocation Delay represents the time elapsed between the initiation of a function invocation and its actual execution. By calculating the mean Invocation Delay for each IP address, we gain insights into the typical processing latency experienced by functions triggered from devices associated with that IP.

4. **InvocationDelay_std**: Similar to RTT_std, InvocationDelay_std represents the standard deviation of Invocation Delay values across instances associated with the same IP address. A higher standard deviation implies greater variability in processing latencies, which could be influenced by factors such as varying workload demands or resource availability.

These statistical features enable us to characterize the behavior of network attributes within specific groups, facilitating the identification of patterns, anomalies, or performance trends associated with different IP addresses. By understanding the distribution and variability of these attributes, network administrators can make informed decisions regarding resource allocation, performance optimization, and anomaly detection.


```python
# 3. Statistical features
X['RTT_mean'] = dataset.groupby('IP')['RTT'].transform('mean')
X['RTT_std'] = dataset.groupby('IP')['RTT'].transform('std')
X['InvocationDelay_mean'] = dataset.groupby('IP')['InvocationDelay'].transform('mean')
X['InvocationDelay_std'] = dataset.groupby('IP')['InvocationDelay'].transform('std')
```

### Entropy-based Features

Entropy-based features offer insights into the uncertainty or randomness present in specific attributes across different instances or categories. These features are computed based on the concept of entropy, which measures the level of unpredictability or disorder within a dataset. In the context of network traffic analysis, entropy-based features provide valuable information about the diversity and distribution of certain attributes across unique IP addresses.

For each of the selected attributes, namely SubmitTime, RTT (Round-Trip Time), ActiveFunctionsAtRequest, and ActiveFunctionsAtResponse, the entropy-based features are calculated. These features capture the degree of variability or dispersion in attribute values within each group of instances associated with a particular IP address.

1. **SubmitTime_entropy**: SubmitTime represents the timestamp when a request or function invocation is submitted. The SubmitTime_entropy feature quantifies the uncertainty or variability in the distribution of submission times across instances associated with each IP address. Higher entropy values suggest a wider range of RTT values or more diverse communication latencies experienced by devices linked to the same IP address.

3. **ActiveFunctionsAtRequest_entropy**: ActiveFunctionsAtRequest represents the number of active functions at the time of request submission. The ActiveFunctionsAtRequest_entropy feature assesses the uncertainty or variability in the distribution of active function counts across instances associated with each IP address. Elevated entropy values indicate greater diversity in the number of active functions present during request submissions, which may reflect varying workload intensities or resource usage patterns.

4. **ActiveFunctionsAtResponse_entropy**: Similar to ActiveFunctionsAtRequest_entropy, ActiveFunctionsAtResponse_entropy quantifies the unpredictability or variability in the distribution of active function counts at the time of response. This feature provides insights into the diversity of response behavior across instances associated with each IP address. Higher entropy values suggest a wider range of active function counts at response times, reflecting the dynamic nature of workload processing or system interactions.

By incorporating entropy-based features into network traffic analysis, organizations can gain deeper insights into the diversity, variability, and randomness present in specific attributes across different IP addresses. These features enable the detection of anomalous patterns, the identification of diverse usage behaviors, and the characterization of network dynamics, ultimately supporting more informed decision-making in network management and optimization.

The preprocessing steps for numerical data involved imputing missing values and scaling the features using Min-Max scaling. Here's a summary of the preprocessing steps:

## Preprocessing Numerical Data

- **Function Trigger Encoding:** The 'functionTrigger' column was encoded using a mapping dictionary to convert categorical values into numerical representations. For example, 'notification' was mapped to 0 and 'other_value' to 1.

- **Imputation:** Missing values in numerical columns were imputed using a constant value strategy. NaN values were replaced with 0 to ensure consistency in feature representation.

- **Scaling:** Min-Max scaling was applied to normalize the numerical features within a specified range. This transformation ensures that all features have the same scale and are comparable across different ranges of values.

These preprocessing steps are essential for preparing the numerical data for machine learning models, ensuring that the data is in a suitable format for training and evaluation.


```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer

function_trigger_mapping = {'notification': 0, 'other_value': 1}

dataset['functionTrigger'] = dataset['functionTrigger'].map(function_trigger_mapping)

numerical_columns = ['Id', 'bot', 'FunctionId', 'functionTrigger', 'SubmitTime', 'RTT',
                     'InvocationDelay', 'ResponseDelay', 'FunctionDuration', 'ActiveFunctionsAtRequest',
                     'ActiveFunctionsAtResponse', 'maxcpu', 'avgcpu', 'p95maxcpu', 'vmcorecountbucket',
                     'vmmemorybucket']

# Impute NaN values with a specific value, such as 0
imputer = SimpleImputer(strategy='constant', fill_value=0)
dataset[numerical_columns] = imputer.fit_transform(dataset[numerical_columns])

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(dataset[numerical_columns])
```

### Data Splitting for Model Training and Evaluation

When preparing data for training machine learning models, it's essential to split the dataset into separate subsets for training, validation, and testing. The commonly used approach is the 80-20 split, where 80% of the data is used for training and validation, and 20% is reserved for testing the model's performance.

Here's how the data splitting is typically conducted:

1. **Initial Splitting:**
   - The dataset is first divided into a training set (80%) and a testing set (20%) using the `test_size=0.2` parameter in the `train_test_split` function.

2. **Further Splitting for Validation:**
   - The training set obtained from the initial split is further divided into a training subset (80% of the training set) and a validation subset (20% of the training set).
   - This results in approximately 64% of the original data being allocated for training, 16% for validation, and 20% for testing.

This approach ensures that the model is trained on a sufficiently large dataset while also having separate datasets for validating its performance during training and evaluating its generalization on unseen data.


```python
# Split the dataset into training, validation, and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

The reshaping of input data is crucial for feeding it into a Long Short-Term Memory (LSTM) neural network, which requires a specific input shape. Here's a summary of the reshaping process:

## Reshaping Input Data for LSTM

- **Timesteps:** Each sample in the input data represents a single timestep. In this case, the `timesteps` variable is set to 1, indicating that each sample has only one timestep.

- **Number of Features:** The number of features in the input data is determined by the shape of the scaled dataset (`X_scaled`). This value is used to ensure compatibility with the LSTM input shape.

- **Reshaping Training, Validation, and Testing Data:** The training (`X_train`), validation (`X_val`), and testing (`X_test`) datasets are reshaped to match the LSTM input shape. The reshaping operation transforms the data into a three-dimensional array, with dimensions corresponding to the number of samples, timesteps, and features.

By reshaping the input data in this manner, it becomes compatible with the LSTM model architecture, enabling efficient training and evaluation.


```python
# Reshape the input data to match the LSTM input shape
timesteps = 1  # Assuming each sample represents a single timestep
num_features = X_scaled.shape[1]  # Use the scaled dataset for reshaping
X_train = X_train.reshape(X_train.shape[0], timesteps, num_features)
X_val = X_val.reshape(X_val.shape[0], timesteps, num_features)
X_test = X_test.reshape(X_test.shape[0], timesteps, num_features)
```

To create and train a bidirectional Long Short-Term Memory (LSTM) model, the following steps are undertaken:

## Creating and Training the Bi-LSTM Model

- **Model Architecture:** The model architecture is defined using the Keras `Sequential` API. The architecture consists of a bidirectional LSTM layer with 64 units and ReLU activation function. The input shape is specified to match the reshaped input data, which includes the number of timesteps and features.

- **Output Layer:** A dense layer with a single unit and sigmoid activation function is added as the output layer. This layer produces binary classification predictions.

- **Compilation:** The model is compiled with the Adam optimizer and binary cross-entropy loss function. Additionally, accuracy is used as the evaluation metric.

- **Early Stopping:** Early stopping is implemented with a patience of 3 epochs to prevent overfitting. The training process monitors the validation loss, and training stops if the validation loss does not improve for three consecutive epochs.

- **Training:** The model is trained using the `fit` method. The training data (`X_train` and `y_train`) are provided, along with the validation data (`X_val` and `y_val`). The training is conducted for 10 epochs with a batch size of 32.

By following these steps, the bidirectional LSTM model is created, compiled, and trained on the provided data, with early stopping implemented to prevent overfitting.


```python
# Create and train the bi-LSTM model
model = Sequential()
model.add(Bidirectional(LSTM(units=64, activation='relu'), input_shape=(timesteps, num_features)))
model.add(Dense(units=1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
early_stopping = EarlyStopping(patience=3, monitor='val_loss')
model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=10, batch_size=32, callbacks=[early_stopping])
```

To assess the performance of the trained bidirectional LSTM model, the following steps are undertaken:

## Assessing Model Performance

- **Training History:** The training history of the model, which includes loss and accuracy metrics on both the training and validation sets across epochs, is obtained.

- **Visualizing Loss:** The loss values during training and validation are plotted against the number of epochs. This visualization helps to assess how the loss changes over the course of training, indicating whether the model is improving or overfitting.

- **Visualizing Accuracy:** Similarly, the accuracy values during training and validation are plotted against the number of epochs. This visualization provides insights into how the model's accuracy evolves during training and validation.

By examining these visualizations, one can gain an understanding of how well the model is learning from the data and whether any overfitting or underfitting is occurring.



```python
# Get training history 
history = model.history 

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper right')
plt.show()

plt.plot(history.history['accuracy']) 
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```

After training the bidirectional LSTM model, predictions are made on the test set to evaluate its performance. The following steps are undertaken:

## Making Predictions

- **Prediction:** The model is used to predict the binary labels for the test set samples.

- **Thresholding:** The predicted probabilities are thresholded at 0.5 to obtain binary predictions. Probabilities above 0.5 are classified as 1 (indicating the presence of a DoW attack), while probabilities below 0.5 are classified as 0 (indicating no DoW attack).

These binary predictions are then used to assess the model's performance in detecting DoW attacks on the test set.



```python
# Make predictions on the test set
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5).astype(int)  # Convert probabilities to binary predictions (0 or 1)
```

To assess the performance of the bidirectional LSTM model in detecting Denial-of-Service (DoW) attacks on the test set, the following steps are performed:

## Evaluation Metrics Calculation

- **Accuracy:** The overall accuracy of the model's predictions on the test set is calculated using the `accuracy_score` function.

- **Confusion Matrix:** A confusion matrix is generated using the `confusion_matrix` function. From the confusion matrix:
  - **True Positives (TP):** The number of instances correctly predicted as DoW attacks.
  - **Actual DoW Attacks:** The total number of instances labeled as DoW attacks in the test set.
  - **Total Samples:** The total number of samples in the test set.

- **Detection Rate (Recall):** The detection rate, also known as recall, is calculated as the ratio of true positives to the total number of actual DoW attacks.

## Results Display

- The calculated accuracy and detection rate are displayed along with the total number of samples, actual DoW attacks, and detected DoW attacks in the test set.



```python
# Assess the presence of DoW attacks
# Assuming y_test and y_pred are arrays of binary labels representing the ground truth and predictions, respectively
from sklearn.metrics import accuracy_score, confusion_matrix

# Evaluate model performance on the test set
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
true_positives = conf_matrix[1, 1]  # Number of true positives
actual_dow_attacks = conf_matrix[1, :].sum()  # Total number of actual DoW attacks
num_samples = len(y_test)  # Total number of samples

# Calculate detection rate (recall)
detection_rate = true_positives / actual_dow_attacks

# Display evaluation metrics
print(f"Accuracy: {accuracy:.2f}")
print(f"Detection Rate: {detection_rate:.2%}")
print(f"Total samples: {num_samples}")
print(f"Actual DoW attacks: {actual_dow_attacks}")
print(f"Detected DoW attacks: {true_positives}")
```

## Evaluation Metrics

The following evaluation metrics are calculated to assess the performance of the model in detecting Denial-of-Service (DoW) attacks:

- **Precision:** Precision measures the proportion of true positive predictions among all positive predictions made by the model.
- **Recall:** Recall, also known as sensitivity, measures the proportion of true positive predictions among all actual positive instances in the test set.
- **F1 Score:** The F1 score is the harmonic mean of precision and recall, providing a single metric to evaluate the balance between precision and recall.
- **ROC AUC Score:** The Receiver Operating Characteristic Area Under the Curve (ROC AUC) score measures the model's ability to distinguish between positive and negative instances across different thresholds.

The calculated scores provide insights into the model's performance in terms of accuracy, completeness, and discrimination capability.



```python
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, roc_curve

# Calculate evaluation metrics
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)

print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
print(f"ROC AUC Score: {roc_auc:.2f}")
```

## Receiver Operating Characteristic (ROC) Curve

The ROC curve visualizes the performance of a binary classification model across different threshold settings. It plots the True Positive Rate (sensitivity) against the False Positive Rate (1 - specificity) at various threshold values.

- **True Positive Rate (TPR):** The proportion of true positive predictions among all actual positive instances in the test set.
- **False Positive Rate (FPR):** The proportion of false positive predictions among all actual negative instances in the test set.

The ROC curve provides a graphical representation of the trade-off between sensitivity and specificity. A perfect classifier would have a curve that passes through the top-left corner of the plot, indicating high TPR and low FPR across all threshold values.

The area under the ROC curve (ROC AUC) quantifies the overall performance of the model: a higher ROC AUC score indicates better discrimination ability.



```python
# Plot the ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
plt.plot(fpr, tpr, label='ROC Curve')
plt.plot([0, 1], [0, 1], 'k--', label='Random')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```
